```python
import numpy as np
import math
import matplotlib.pyplot as plt

def normal_density(mean, variance, x):
    return (1 / math.sqrt(2 * np.pi * variance)) * np.exp(-((x - mean) ** 2) / (2 * variance))

def integration(mean, variance, a,b):
    n = 1000
    dx = (b - a)/n
    midpoints = np.linspace(a + dx/2, b - dx/2, n)
    total_area = np.sum(normal_density(mean, variance, midpoints) * dx)
    return total_area
    
def average_dosage(height):
    return height * height * 2.38
if __name__ == '__main__':
    mean = 171
    variance = 7.1
    
    x_values = np.linspace(0, 200, 100)
    print("Probability of between 162cm to 190cm:", integration(mean, variance, 162, 190))
    print("Average dosage:", average_dosage(171))
    y_values = normal_density(mean, variance, x_values)
    plt.figure(figsize=(8, 5))
    plt.plot(x_values, y_values, label=f'Normal Distribution (mean={mean}, variance={variance})')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Normal Distribution Curve')
    plt.legend()
    plt.grid()
    plt.show()

    
```
