```python
#(a)
x, y = sp.symbols('x y')
f = sp.exp(x) * sp.sin(y) + y**3
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

df_dx, df_dy
```

```python
#(b)
x, y = sp.symbols('x y')
g = x**2 * y + x * y**2
gradient_g = [sp.diff(g, var) for var in (x, y)]
point = {x: 1, y: -1}
gradient_at_point = [expr.subs(point) for expr in gradient_g]
magnitude = sp.sqrt(sum(comp**2 for comp in gradient_at_point))

gradient_at_point, magnitude
```

```python
#(c)
x, y = sp.symbols('x y')
h = sp.ln(x**2 + y**2)
dh_dx = sp.diff(h, x)
dh_dy = sp.diff(h, y)

# ∂²h/∂x²
d2h_dx2 = sp.diff(dh_dx, x) 
 # ∂²h/∂y²
d2h_dy2 = sp.diff(dh_dy, y) 
 # ∂²h/∂x∂y
d2h_dxdy = sp.diff(dh_dx, y)

d2h_dx2, d2h_dy2, d2h_dxdy

#I the function h(x,y) and its second order mixed partial derivatives are continuous, 
#then ∂²h/∂x∂y= ∂²h/∂y∂x, and this equals -4*x*y/(x**2 + y**2)**2 which means the mixed
#partial derivatives are equal and therefore symmetrical
```

```python
#(d)
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
